## 폐암 수술 환자의 생존율 예측

* 데이터를 불러오기 위해 pandas 라이브러리를 사용

```
import io
import pandas as pd
```

* 데이터 불러오기

```
data_pandas = pd.read_csv('ThoraricSurgery.csv',header=None)
```

```
data_pandas
```

* 케라스 함수 호출하고 numpy 불러오기

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf
import numpy as np
```

* 실행할 때마다 같은 결과를 출력하기 위해서 seed값 설정

```
np.random.seed(3)
tf.random.set_seed(3)
```

* 속성인 환자의 기록을 X, 클래스인 수술 결과를 Y로 저장
* 인덱싱 : 그 데이터를 가리키는 것

  -- loc : 인덱스 이름으로 직접 참조

  -- iloc : 정수 인덱스값

* 슬라이싱 : 데이터를 자르는 것

```
#[행, 열]
#[시작점:끝점, 시작점:끝점]
# : 앞뒤에 아무것도 적지 않으면 '다'라는 뜻
X=data_pandas.iloc[:,:17]
Y=data_pandas.iloc[:,17]
X
```

```
Y
```

* 모델 구조 결정 (입력층, 은닉층, 출력층)

```
model=Sequential()
model.add(Dense(30, input_dim=17, activation='relu')) #은닉층
model.add(Dense(1, activation='sigmoid')) #출력층
```

*!!이제 모델을 실행해보자
!!*

mean_squared_error 와 binary_crossentropy 의 특징을 이해하고 실행 결과를 예측 및 비교

```
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
#model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, Y, epochs=100, batch_size=10)
```
